*Exercise 4:* Write a function that takes a `n` by `n` matrix representing a tic-tac-toe game, and returns -1, 0, 1, or 2 indicating the game is incomplete, the game is a draw, player 1 has won, or player 2 has one, respectively. Here are some example inputs you can use to test your code:

````python
def check_tic_tac_toe(matrix):
    n = len(matrix)

    # Check rows and columns for a win
    for i in range(n):
        if all(matrix[i][j] == 1 for j in range(n)):  # Check row for player 1
            return 1
        if all(matrix[i][j] == 2 for j in range(n)):  # Check row for player 2
            return 2
        if all(matrix[j][i] == 1 for j in range(n)):  # Check column for player 1
            return 1
        if all(matrix[j][i] == 2 for j in range(n)):  # Check column for player 2
            return 2

    # Check diagonals for a win
    if all(matrix[i][i] == 1 for i in range(n)):  # Check main diagonal for player 1
        return 1
    if all(matrix[i][i] == 2 for i in range(n)):  # Check main diagonal for player 2
        return 2
    if all(matrix[i][n - 1 - i] == 1 for i in range(n)):  # Check anti-diagonal for player 1
        return 1
    if all(matrix[i][n - 1 - i] == 2 for i in range(n)):  # Check anti-diagonal for player 2
        return 2

    # Check for any empty cells
    for row in matrix:
        if 0 in row:
            return -1  # Game is incomplete

    return 0  # Game is a draw

# Example usage:
matrix1 = [
    [1, 1, 1],
    [0, 2, 2],
    [0, 0, 0]
]  # Player 1 wins

matrix2 = [
    [1, 2, 1],
    [2, 1, 2],
    [2, 1, 1]
]  # Draw

matrix3 = [
    [1, 2, 0],
    [0, 1, 2],
    [0, 0, 0]
]  # Game is incomplete

print(check_tic_tac_toe(matrix1))  # Output: 1
print(check_tic_tac_toe(matrix2))  # Output: 0
print(check_tic_tac_toe(matrix3))  # Output: -1
````

*Exercise 9:* Use all of the previous exercises to implement a full tic-tac-toe game, where an appropriate board is drawn, 2 players are repeatedly asked for a location coordinates of where they wish to place a mark, and the game status is checked until a player wins or a draw occurs.

````python
def convert_coordinates(location):
    """Convert a location string like 'A1' to numerical coordinates (x, y)."""
    row = ord(location[0].upper()) - ord('A')  # Convert row letter to index (0-based)
    col = int(location[1]) - 1  # Convert column number to index (0-based)
    return row, col

def place_marker(board, player, coordinates):
    """Place a marker for the player at the specified coordinates on the board."""
    x, y = coordinates

    if 0 <= x < len(board) and 0 <= y < len(board[0]) and board[x][y] == 0:
        board[x][y] = player  # Place the player's marker
        return True  # Successful placement
    return False  # Failed placement

def make_move(board, player, location):
    """Make a move for the player at the specified location in 'Ax' format."""
    coordinates = convert_coordinates(location)  # Convert location to coordinates
    return place_marker(board, player, coordinates)  # Place the marker

def print_board(board):
    """Print the current state of the board with labels."""
    print("   A   B   C")
    for i, row in enumerate(board):
        print("+---+---+---+")
        print(f"{i + 1}|", end="")
        for cell in row:
            if cell == 0:
                print("   |", end="")
            elif cell == 1:
                print(" X |", end="")
            elif cell == 2:
                print(" O |", end="")
        print()
    print("+---+---+---+")

def check_winner(board):
    """Check for a winner on the board."""
    # Check rows, columns, and diagonals
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != 0:  # Check rows
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != 0:  # Check columns
            return board[0][i]
    
    if board[0][0] == board[1][1] == board[2][2] != 0:  # Check diagonal
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != 0:  # Check anti-diagonal
        return board[0][2]
    
    return 0  # No winner yet

def is_draw(board):
    """Check if the game is a draw."""
    return all(cell != 0 for row in board for cell in row)

def get_valid_move(board, player):
    """Prompt the player for a valid move until a valid input is received."""
    while True:
        location = input(f"Player {player}, enter your move (e.g., A1): ")
        
        if len(location) != 2 or not location[0].isalpha() or not location[1].isdigit():
            print("Invalid input. Please use the format 'Ax' where x is a number.")
            continue
        
        result = make_move(board, player, location)
        
        if result:
            print(f"Player {player} placed their marker at {location}.")
            break
        else:
            print("Invalid move. Please try again.")

def play_game():
    """Main function to play the Tic Tac Toe game."""
    board = [[0, 0, 0] for _ in range(3)]  # Initialize the board
    current_player = 1  # Player 1 starts the game
    
    while True:
        print_board(board)  # Display the current board
        get_valid_move(board, current_player)  # Get a valid move from the current player
        
        # Check for a winner
        winner = check_winner(board)
        if winner:
            print_board(board)  # Display the final board
            print(f"Player {winner} wins!")
            break
        
        # Check for a draw
        if is_draw(board):
            print_board(board)  # Display the final board
            print("It's a draw!")
            break
        
        # Switch players
        current_player = 2 if current_player == 1 else 1

# Start the game
play_game()
````
